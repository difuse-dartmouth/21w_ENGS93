```r
n=20
se0=40/sqrt(n)
xcrit=25 
curve(pnorm((xcrit-x)/se0),from=mu0- 4*se0,to=mu1+4*se0,xlab="mu",ylab="OC(mu)") 
abline(v=15)
abline(v=25) 
abline(v=40)
```

### Question 1 part d)

```r
n=20
se0=40/sqrt(n)
xcrit=25
curve(pnorm((xcrit-x)/se0),from=mu0-4*se0, to=mu1+4*se0, xlab="mu", ylab="OC(mu)")
xcrit=35
curve(pnorm((xcrit-x)/se0), col="red", add=TRUE)
abline(v=15)
abline(v=25)
abline(v=40)
```

The earlier values were approximately 0.85, 0.5, 0.05. Now they are approximately 1, 0.9, 0.3. So they have all increased. When the rejection region is X_bar>35, the OC curve shifts to the right. Note that this decreases alpha values and increases beta values. In other words, both 1-alpha and beta increase, you are less likely to reject the null hypothesis since the rejection region has shrunk in size.
